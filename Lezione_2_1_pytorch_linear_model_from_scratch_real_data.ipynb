```python
#Build a dummy dataset
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
diabetes.target[:3]
diabetes.data.shape
```

```python
print(diabetes.DESCR)
```

```python
# Load the dataset
diabetes_data = load_diabetes()
df = pd.DataFrame(data=diabetes_data.data, columns=diabetes_data.feature_names)

# Add the target column to the DataFrame
df['progression'] = diabetes_data.target
```

```python
import matplotlib.pyplot as plt

# Histogram of the progression
df['progression'].hist(bins=30)
plt.title('Histogram of Diabetes Progression')
plt.xlabel('Progression')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns

# Select a subset of features to avoid a cluttered pair plot
subset_df = df[['age', 'bmi', 'bp', 'progression']]

# Pair plot
sns.pairplot(subset_df)
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Diabetes Dataset Features')
plt.show()
```
